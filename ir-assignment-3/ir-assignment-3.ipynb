# Assignment 3 - Textblob for Sentiment Analysis

TextBlob’s output
<br/>
<br/><b>polarity</b> is a float value within the range [-1.0, 1.0] 
<br/>where
<br/>-1.0 is a negative polarity
<br/>1.0 is positive. 
<br/>0 stands for a neutral evaluation of a statement as it doesn’t contain any words from the training set.
<br/>
<br/><b>subjectivity/objectivity</b> identification task reports a float within the range [0.0, 1.0] 
<br/>where 
<br/>0.0 is a very objective sentence 
<br/>1.0 is a very subjective sentence.

```python
from textblob import TextBlob
```

```python
fileName = 'assignment-3-data.txt'
```

```python
file = open(fileName, 'r')
Lines = file.readlines()
```

```python
for line in Lines:
    print('─' * 60)  
    print(line.strip())
    txt_blb_score = TextBlob(line.strip())
    print('Polarity: {:.2f}, Subjectivity: {:.2f}'.format(txt_blb_score.sentiment.polarity, txt_blb_score.sentiment.subjectivity))
    
    if txt_blb_score.sentiment.polarity < 0 :
        print('Polarity: Negative sentiment')
    if txt_blb_score.sentiment.polarity == 0:
        print('Polarity: Neutral sentiment')
    if txt_blb_score.sentiment.polarity > 0:
        print('Polarity: Positive sentiment')
        
    if txt_blb_score.sentiment.subjectivity <= 0.30:
        print('Subjectivity: Fact based statement')
    else:
        if txt_blb_score.sentiment.subjectivity >= 0.70:
            print('Subjectivity: Opinion based statement')
        else: 
            print('Subjectivity: Generic statement')
```
